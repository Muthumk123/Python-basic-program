```python
#1. Scenario: A user is required to enter a valid number in a form, but users
#sometimes input invalid data.
#Write logic to repeatedly prompt the user until they enter a valid integer.
def validation(textfield):
        if textfield.isdigit():
            num = int(textfield)
            if num <= 8:
                print("Valid:", num)
                   
            else:
                print("Invalid! Number must be at least 5.")
        else:
            print("Invalid input! Please enter digits only.")

        
    
```

```python
#Scenario: A data analysis tool processes a list of numbers and needs to identify
#the most frequently occurring value.
#Write logic to find the most frequently occurring number in a given list.
def number_repet():
    num=[1,2,2,3,3,3,5,6,7,8,9,2,5,55,45]
    occur=max(num,key=num.count)
    print("Most occurring number is:", occur)
    
   
```

```python
#3.	Scenario: A text-processing application needs to compare words and check if they are anagrams (contain the same letters in a different order).
# Write logic to determine whether two given strings are anagrams.
def aragram(str1,str2):
    text1= str1.replace("" , "").lower()
    text2= str2.replace("" , "").lower()
    sort1="".join(sorted(text1))
    sort2="".join(sorted(text2))
    if sort1==sort2:
            print("The string was anagrams",sort1,sort2)
    else:
        print("The string was not anagrams")
    


```

```python
aragram("listen", "silent")
```

```python
#4.	Scenario: A speech analysis program needs to count the number of vowel sounds in a given input.
# Write logic to count the number of vowels in a given string.
def vowels(word):
    vol= "aeiouAEIOU"
    count=0
    for ch in word:
           if ch in vol:
                 count+=1        
    print("The count of vowels is:",count)

```

```python
vowels("education")
```

```python
#5.	Scenario: A text-editing software includes a feature to reverse the order of words in a sentence for stylistic effects.
 #Write logic to reverse the order of words in a sentence while keeping the words themselves intact.
def reverse(sentence):
    word=sentence.split()
    reves_word=word[::-1]
    return "".join(reves_word)
    
    
```

```python
reverse("Here is the simple and clean logic to reverse the order of words in a sentence while keeping each word unchanged")
```

```python
#6.	Scenario: A missing number is detected in a sequence of values stored in a database.
 #Write logic to find the missing number in a list containing n-1 numbers from 1 to n.

def find_missing_simple():
    num_list = [1, 2, 4, 5, 6]
    my_list=len(num_list)+1
    expected=0
    for i in range(1,my_list+1):
        expected+=i

        
    actual=sum(num_list)
    return expected -actual
        
```

```python

print("Missing number is:", find_missing_simple())
```

```python
#10.Scenario: A language-learning app wants to verify whether a given sentence is a pangram (contains every letter of the alphabet at least once).
def is_pangram(sentence):
    sentence = sentence.replace(" ", "").lower()
    return set("abcdefghijklmnopqrstuvwxyz").issubset(set(sentence))
    
print(is_pangram("The quick brown fox jumps over the lazy dog"))
```

```python
#3.Find Even and Odd Numbers in a List
#Question: Write a Python program to separate even and odd numbers from the list [12, 17, 19, 24, 35, 40, 56].
def separate_odd_even():
    num=[12, 17, 19, 24, 35, 40, 56]
    for i in num:
         if i%2==0:
             print("Even=",i)
         else:
             print("Odd=",i)
```
