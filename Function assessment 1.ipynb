```python
def Eligible(Gender, age):
    print("Your Gender:", Gender)
    print("Your Age:", age)
    if Gender == "male" and age > 21:
        print("Eligible for marriage:")
        Marriage = "Eligible"
    elif Gender == "female" and age > 18:
        print("Eligible for marriage:")
        Marriage = "Eligible"
    else:
        print("Not Eligible")
        Marriage = "Not Eligible"
    return Marriage

        
        
        
        
    
```

```python
Eligible("male", 20)
```

```python
def marks_report(s1, s2, s3, s4, s5):
    total = s1 + s2 + s3 + s4 + s5
    percentage = (total / 500) * 100
    print("Subject1 =", s1)
    print("Subject2 =", s2)
    print("Subject3 =", s3)
    print("Subject4 =", s4)
    print("Subject5 =", s5)
    print("Total :", total)
    print("Percentage :", percentage)
```

```python
marks_report(98, 87, 95, 95, 93)
```
